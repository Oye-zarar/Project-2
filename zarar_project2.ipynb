```python
import os  # Module to interact with the file system
import PyPDF2  # Library for working with PDFs
# Path to the folder containing PDFs
folder_path = r'C:\\Users\\ZARAR\\Desktop\\ML Python\\Projects\\project 2'
pdf_texts = []  # List to store extracted text from each PDF
# Walk through the folder and its subfolders
for root, dirs, files in os.walk(folder_path):
    for file in files:
        # Check if the file is a PDF
        if file.endswith('.pdf'):
            # Get the full file path
            file_path = os.path.join(root, file)
            
            # Open the PDF in read-binary mode
            with open(file_path, 'rb') as f:
                reader = PyPDF2.PdfReader(f)  # Create PDF reader object
                text = ''  # Initialize empty string for text
                
                # Loop through each page in the PDF and extract text
                for page in reader.pages:
                    text += page.extract_text()  # Add extracted text from the page
                
                # Add the extracted text to the list
                pdf_texts.append(text)
                
                # Print the first 100 characters of the extracted text
                print(f"Extracted from {file_path}:\n", text[:100])  
```

**	spaCy with NER for identifying entities**

```python
import spacy
nlp = spacy.load('en_core_web_sm')
for text in pdf_texts:
    doc = nlp(text)
    
    # Print the recognized entities for each PDF
    for ent in doc.ents:
        print(ent.text, ent.label_)
```

**Job Description Parsing**

```python
# Sample job description text
job_desc_text = """
We are looking for a data scientist with experience in Python, machine learning, and data analysis. 
The candidate should have a degree in Computer Science or related field, and at least 3 years of experience in a similar role.
"""
doc = nlp(job_desc_text)
print("Entities found in the job description:")
for ent in doc.ents:
    print(f"{ent.text} ({ent.label_})")
```

**Extract Keywords Using TF-IDF**

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Sample job descriptions
job_descs = [
    "We are looking for a data scientist with experience in Python, machine learning, and data analysis.",
    "The candidate should have a degree in Computer Science or related field, and at least 3 years of experience in a similar role.",
    "Looking for a software engineer proficient in Java, web development, and agile methodologies."
]

# Create TF-IDF vectorizer
vectorizer = TfidfVectorizer(stop_words='english')

# Fit and transform job descriptions
X = vectorizer.fit_transform(job_descs)

# Get feature names
feature_names = vectorizer.get_feature_names_out()

# Print keywords and their TF-IDF scores
for i, job_desc in enumerate(job_descs):
    print(f"\nKeywords for Job Description {i+1}:")
    scores = X[i, :].toarray()[0]
    keywords = [(feature_names[j], scores[j]) for j in range(len(scores))]
    sorted_keywords = sorted(keywords, key=lambda x: x[1], reverse=True)
    for keyword, score in sorted_keywords:
        print(f"{keyword}: {score:.4f}")
```
